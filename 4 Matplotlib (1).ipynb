### 4. Изучите документацию matplotlib и попробуйте построить на одном изображении 4 системы координат, на верхних двух - квадратичную и кубическую параболы, на нижних - экспоненту и логарифм.


```python
n = np.linspace(1,20,10)
```

```python
quatro = [x**2 for x in n]
cube = [x**3 for x in n]
expon = [math.e**x for x in n]
logariphm = [math.log1p(x) for x in n]
```

```python


plt.subplot(221)
plt.plot(quatro)

plt.subplot(222)
plt.plot(cube)

plt.subplot(223)
plt.plot(expon)

plt.subplot(224)
plt.plot(logariphm)
```

### 5. Изучите вопрос построения heatmap и изобразите его для функцииf(x, y) = 3xy + x − 2y в области [0; 5]2
.

```python
def func(x,y):
    return (3*x*y)+x-2*y
```

```python
% matplotlib inline
from matplotlib import pylab as plt
import pandas as pd
import seaborn as sns
x = []
y = []
f = []

for i in range(6):
    for j in range(6):
        x.append(i)
        y.append(j)
        f.append(func(i,j))
data = pd.DataFrame()
data['x'] = x
data['y'] = y
data['f'] = f
#print(data)

plot = data.pivot(index = 'x', columns = 'y', values = 'f')
sns.heatmap(plot, annot=True, fmt="g")
```
