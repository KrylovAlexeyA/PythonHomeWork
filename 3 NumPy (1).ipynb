## 2. Попробуйте сравнить по производительности numpy.array и list, выполняя какую-то стандартную операцию с каждым из типов много раз.


```python
a = np.arange(0,100_000)
l = [i for i in range(100_000)]
```

```python
import time
start_time = time.time()
a = a**2
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
import time
start_time = time.time()
l = [l[i]**2 for i in range(len(l))]
print("--- %s seconds ---" % (time.time() - start_time))
```

### numpy array намного быстрее

## 3. Изучите numpy.linspace(), напишите свою реализацию аналогичной функции с помощью list comprehension, дающий тот же результат, только в виде list, а не numpy.ndarray. Сравните по производительности два варианта на достаточно больших массивах.

```python
a = np.linspace(1,100_000, 10000)
```

```python
a
```

```python
def linspace(start,stop,num):
    step = float(stop-start)/float(num)
    list = [start + step*i for i in range(0,stop)]
    return list
    
```

```python
linspace(1,100_000,10000)
```

```python
import time
start_time = time.time()
a = np.linspace(1,100_000, 100)
print("--- %s seconds ---" % (time.time() - start_time))
```

```python
import time
start_time = time.time()
linspace(1,100_000,100)
print("--- %s seconds ---" % (time.time() - start_time))
```

### numpy опять быстрее
