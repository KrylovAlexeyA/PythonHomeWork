## 7. Изучите примеры работы с декораторами из официальной документации. Фактически, декоратор - это функция, которая принимает на вход другую функцию и возвращает её в модифицированном виде. Синтаксически предусмотрена возможность использовать декоратор, написав перед объявлением функции имя_декоратора. Попробуйте найти реализацию (или сделать сами) кэширующего декоратора и декоратора, вычисляющего время работы функции. Попробуйте применить эти декораторы в разной последовательности к функции, вычисляющей числа Фибоначчи, сравнить и объяснить результаты. Объясните также, как работают использованные вами декораторы.


```python
def memorized(f):
    memory = {}

    def wrapper(*args, **kwargs):
        key = (tuple(args), hash(tuple(sorted(kwargs.items()))))
        if not key in memory:
            memory[key] = f(*args, **kwargs)
        return memory[key]
    return wrapper
```

```python
import time

def timer(f):
    def tmp(*args, **kwargs):
        t = time.time()
        res = f(*args, **kwargs)
        print ("Время выполнения функции %s: %f" % (str(f), (time.time()-t)))
        return res

    return tmp
```

```python
@timer
def fibonachi(n):
    if n > 2:
        return fibonachi(n - 1) + fibonachi(n - 2)
    else:
        return 1
    
```

```python
fibonachi(20)
```

```python
@memorized
@timer
def fibonachi(n):
    if n > 2:
        return fibonachi(n - 1) + fibonachi(n - 2)
    else:
        return 1
```

```python
fibonachi(20)
```

```python

@timer
@memorized
def fibonachi(n):
    if n > 2:
        return fibonachi(n - 1) + fibonachi(n - 2)
    else:
        return 1
```

```python
fibonachi(20)
```

### 9. Напишите setter и getter для полей класса комплексного числа.

```python
class ComplexNumber:
    def __init__(self,x,y):
        self._x = x
        self._y = y
    def getModule(self):
        return abs(self._x + self._y )
    def getArgument(self):
        return math.atan(self._y /self._x)
    def setX(self,x):
        self._x = x
    def setY(self, y):
        self._y = y
    def getX(self):
        return self._x
    def getY(self):
        return self._y   
```

```python
num =  ComplexNumber(5,7)
```

```python
num.getX()
num.getY()
```

```python
num.setX(2)
num.setY(8)
```
